```python
import cv2
import time
```

```python
#Give the path to video file or numbers will be using the webcam to capture video
video=cv2.VideoCapture(0) #0-Built in camera,1-one webcam 2-two webcam
a=1
while True:
    a=a+1
    
    check,frame=video.read()#Read the first frame/image of the video
#check is a boolean data type and it returns true if python is able to read the video capture object
#frame is a numpy array,it represents the first image that video captures
    print(check)
    print(frame)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('CaptureVideo',gray)
    key = cv2.waitKey(1)
    if key==ord('q'):
         break
print(a)
#video.release()
cv2.destroyAllWindows()
```
